```python
import pandas as pd
datasets_of_interest = ['FICO', 'CKD', 'HEART_DISEASE', 'BREAST_CANCER', 'MIMIC', 'ADULT', 'PHARYNGITIS']
imputations_of_interest = ['Mean', 'MICE', 'MissForest', 'MIWAE']

load_statuses_baselines = {
    'dataset': [],
    'imputation': [],
    'slurm_iter': [],
    'success': []
}
for dataset_of_interest in datasets_of_interest:
    for ds_name in [f'{dataset_of_interest}', f'{dataset_of_interest}_0.25', f'{dataset_of_interest}_0.5', f'{dataset_of_interest}_0.75']:
        for imputation_method in imputations_of_interest:
            for s_iter in range(120):
                load_statuses_baselines['slurm_iter'] = load_statuses_baselines['slurm_iter'] + [s_iter]
                load_statuses_baselines['imputation'] = load_statuses_baselines['imputation'] + [imputation_method]
                load_statuses_baselines['dataset'] = load_statuses_baselines['dataset'] + [ds_name]
                
                #df = pd.read_csv(f'./parallelized_results/baselines_2024-02_01_iter_{s_iter}_10_imp_all.csv')
                try:
                    df = pd.read_csv(f'./parallelized_results/baselines_iter_{s_iter}_{ds_name}_{imputation_method}.csv')
                    load_statuses_baselines['success'] = load_statuses_baselines['success'] + [1]
                except:
                    load_statuses_baselines['success'] = load_statuses_baselines['success'] + [0]
```

```python
load_statuses_baselines = pd.DataFrame(load_statuses_baselines)
load_statuses_baselines['success'].mean()
```

```python

load_statuses_imps = {
    'dataset': [],
    'imputation': [],
    'success': []
}
for dataset_of_interest in datasets_of_interest:
    for ds_name in [dataset_of_interest]:
        for imputation_method in imputations_of_interest:
            for subsample in ['', '_0.25', '_0.5', '_0.75']:
                load_statuses_imps['imputation'] = load_statuses_imps['imputation'] + [imputation_method]
                load_statuses_imps['dataset'] = load_statuses_imps['dataset'] + [ds_name + subsample]

                try:
                    df = pd.read_csv(f'../../handling_missing_data/timing_stats_{ds_name}{subsample}_{imputation_method}_5_3.csv')
                    df = df.drop_duplicates(['imputation', 'dataset', 'holdout_set', 'validation_set', 'm'])
                    df = df[(df['dataset'] == ds_name + subsample) & (df['imputation'] == imputation_method)]
                    load_statuses_imps['success'] = load_statuses_imps['success'] + [df.shape[0] / 500]
                except:
                    load_statuses_imps['success'] = load_statuses_imps['success'] + [0]
```

```python
load_statuses_imps = pd.DataFrame(load_statuses_imps)
load_statuses_imps['success'].mean()
```
